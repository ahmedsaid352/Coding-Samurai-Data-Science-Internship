```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# ignore pandas warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('HousingPrices-Amsterdam-August-2021.csv')
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns = {'Unnamed: 0':'id'})
df.info()
```

```python
df[df['Price'].isna()]
```

```python
df = df.fillna(df.mean())
df.info()
```

```python
df.describe()
```

```python
import plotly.graph_objects as go

# Create a scatter mapbox trace
trace = go.Scattermapbox(
    lat=df.Lat,  #  latitude values
    lon=df.Lon,  #  longitude values
    mode='markers',
    marker=dict(
        size=10,
        opacity=0.7
    ),
)

# Create layout for the map
layout = go.Layout(
    mapbox=dict(
        style='carto-positron',
        zoom=10,
        center=dict(lat=52.364499, lon=4.886818),
    ),
    title='House locations',
)

# Create the figure with the trace and layout
fig = go.Figure(data=[trace], layout=layout)
# Show the plot
fig.show()
```

 we can notice that there are about three or four clusters of house locations

### BlitzML is a Python package that automates machine learning pipelines rapidly (my graduation project)
[For more info](https://github.com/blitzml#clustering)

```python
from blitzml.unsupervised import Clustering


# create the pipeline
auto = Clustering(df[['id','Lon','Lat']], clustering_algorithm = 'auto',feature_selection='importance') # use the K-Mean algorithm

# first perform data preprocessing
auto.preprocess()
# second train the model
auto.train_the_model()

# After training the model we can generate:
auto.gen_pred_df()
auto.gen_metrics_dict()

# We can get their values using:
print(auto.pred_df.head())
print(auto.metrics_dict)
```

```python
clusters_df = auto.pred_df
clusters_df.head()
```

```python
df['cluster'] = clusters_df['cluster'].astype(int)
df.head()
```

```python
df.groupby('cluster')['Price'].mean().plot();
```

```python
# Plot a colored scatter map using Seaborn
sns.scatterplot(data=df, x='Lon', y='Lat', hue='cluster', palette='viridis')

# Set plot labels and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Colored Map')

# Show the plot
plt.show()
```
