### BlitzML is a Python package that automates machine learning pipelines rapidly (my graduation project)
[For more info](https://github.com/blitzml#clustering)

```python
from blitzml.unsupervised import Clustering


# create the pipeline
auto = Clustering(df[['id','Lon','Lat']], clustering_algorithm = 'KM',feature_selection='importance',n_clusters=7) # use the K-Mean algorithm

# first perform data preprocessing
auto.preprocess()
# second train the model
auto.train_the_model()

# After training the model we can generate:
auto.gen_pred_df()
auto.gen_metrics_dict()

# We can get their values using:
print(auto.pred_df.head())
print(auto.metrics_dict)
```

```python
clusters_df = auto.pred_df
clusters_df.head()
```

```python
df['cluster'] = clusters_df['cluster'].astype(int)
df.head()
```

```python
df.groupby('cluster')['Price'].mean().plot();
```

```python
import plotly.express as px


# Plot a colored scatter map using Plotly
fig = px.scatter_mapbox(df, lat='Lat', lon='Lon', color='cluster',
                        color_continuous_scale='Viridis',
                        mapbox_style='carto-positron', zoom=10, center={'lat': 52.364499, 'lon': 4.886818})

# Show the plot
fig.show()
```

```python
df.cluster = df.cluster.astype(str)
df.info()
```

```python
df.drop(columns = ['id', 'Address','Zip'], axis = 1, inplace = True)
df.head()
```

```python
df = pd.concat([df, pd.get_dummies(df['cluster'], prefix='cluster')], axis=1)
df.head()
```

```python
df.drop(columns = ['cluster', 'cluster_0'], axis = 1, inplace = True)
df.head()
```

```python
x_train = df.iloc[:800,:].drop(['Price'],axis = 1)
y_train = df.iloc[:800,:]['Price']
x_test = df.iloc[800:,:].drop(['Price'],axis = 1)
y_test = df.iloc[800:,:]['Price']
```

```python
from sklearn.linear_model import ElasticNet,Lasso,BayesianRidge,SGDRegressor,LinearRegression
from sklearn.ensemble import RandomForestRegressor,  GradientBoostingRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.model_selection import KFold, cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVR
from catboost import CatBoostRegressor
import xgboost as xgb
from xgboost.sklearn import XGBRegressor
```

```python
estimators = {
    "EN":ElasticNet(),
    "lasso":Lasso(),
    "BR":BayesianRidge(),
    "SGD":SGDRegressor(),
    "LR":LinearRegression(),
    "RF":RandomForestRegressor(),
    "GB":GradientBoostingRegressor(),
    "KR":KernelRidge(),
    "SVR":SVR(),
    "catboost":CatBoostRegressor(),
    "xgb":XGBRegressor()
}
```

```python
from sklearn.model_selection import cross_validate

import sys

results = pd.DataFrame(columns=["Regressor", "Avg_r2"])

sys.stdout = open('trash', 'w')

for name, reg in estimators.items():
    model = reg
    cv_results = cross_validate(
        model, x_train , y_train , cv=10,
        scoring=['r2']
    )
    results = results.append({
        "Regressor": name,
        "Avg_r2": cv_results['test_r2'].mean(),
    }, ignore_index=True)
sys.stdout = sys.__stdout__
results = results.sort_values("Avg_r2", ascending=False)
results
```

```python
gb = GradientBoostingRegressor(random_state=1)
rf = RandomForestRegressor(random_state=1)
cb = CatBoostRegressor()
xgb = XGBRegressor()
br = BayesianRidge()
kr = KernelRidge()
lasso = Lasso()

regressors = [
    ('gb',gb),
#     ('rf',rf),
    ('cb',cb),
#     ('br',br),
#     ('kr',kr),
#     ('xgb',xgb),
#     ('lasso',lasso),
    
             ]
```

```python
from sklearn.ensemble import VotingRegressor

model = VotingRegressor(estimators=regressors)

# Redirect output to suppress progress
sys.stdout = open('trash', 'w')

model.fit(x_train, y_train)

# Restore output
sys.stdout = sys.__stdout__
```

```python
y_pred = model.predict(x_test)
```

```python
len(y_test) , len(y_pred)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
